### Taxa de Retorno simples com o Plotly 

```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt 
```

```python
df1 = wb.DataReader('AMZN', data_source = 'yahoo', start = '2012-01-01')
```

```python
df2 = wb.DataReader('AAPL', data_source = 'yahoo', start = '2012-01-01')
```

```python
df3 = wb.DataReader('MSFT', data_source = 'yahoo', start = '2012-01-01')
```

```python
df1.head()
```

```python
df1.reset_index(inplace=True,drop=False)
```

```python
df2.reset_index(inplace=True,drop=False)
```

```python
df2.head()
```

```python
df3.reset_index(inplace=True,drop=False)
```

```python
# Taxa de retorno Simples (Amazon)
```

```python
df1['simple_return'] = (df1['Adj Close'] / df1['Adj Close'].shift(1)) - 1
```

```python
# Plotando a taxa de retorno simples (Amazon)

df1['simple_return'].plot(figsize=(8,5))
plt.show()
```

```python
# Taxa de retorno Simples (Apple)
```

```python
df2['simple_return'] = (df2['Adj Close'] / df2['Adj Close'].shift(1)) - 1
```

```python
#Plotando a taxa de retorno simples (Apple)

df2['simple_return'].plot(figsize=(8,5))
plt.show()
```

```python
# Taxa de retorno Simples (Microsoft)
```

```python
df3['simple_return'] = (df3['Adj Close'] / df3['Adj Close'].shift(1)) 
```

```python
# Plotando a taxa de retorno simples (Microsoft)

df3['simple_return'].plot(figsize=(8,5))
plt.show()
```

```python
import plotly.express as px 
```

```python
fig1 = px.line(df1, x = 'Date', y = 'simple_return', title = 'Amazon Time Series (Rate of Return)')

figure1 = fig1.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="YTD", step="year", stepmode="todate"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all")
        ])
    )
)
fig1.show()
```

```python
fig2 = px.line(df2, x = 'Date', y = 'simple_return', title = 'Apple Time Series (Simple Tax Return)')

figure2 = fig2.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="YTD", step="year", stepmode="todate"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all")
        ])
    )
)
figure2.show()
```

```python
fig3 = px.line(df2, x = 'Date', y = 'simple_return', title = 'Microsoft Time Series (Simple Tax Return)')

figure3 = fig3.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="YTD", step="year", stepmode="todate"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all")
        ])
    )
)
figure3.show()
```
